```python
# Let's test run some descriptives: what does the distribution of customer ratings looks like?

sns.distplot(df['Rating'])

plt.axvline(x = np.mean(df['Rating']), c = 'red', ls = '--', label = 'mean') # axis vertical line is mean

plt.axvline(x = np.percentile(df['Rating'], 25), c = 'green', ls = '--', label = '25th/75th percentile')

plt.axvline(x = np.percentile(df['Rating'], 75), c = 'green', ls = '--', label = '25th/75th percentile')

plt.legend(loc = 'upper right', prop={'size': 5.5})

plt.rcParams['figure.dpi'] = 300
```
